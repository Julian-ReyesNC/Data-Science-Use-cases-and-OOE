```python
import pandas as pd
csvFile = 'Real_Estate_Sales.csv'
reHandle = pd.read_csv(csvFile)
print(reHandle.head(10))
num_data_points = reHandle.shape[0]
column_names = reHandle.columns
statistical_summary = reHandle.describe()
print(f"\n\n")
print(f"The number of data points are {num_data_points} \n\n")
print(f"Here are the column names: \n\n{column_names}\n\n")
print(f"Here is a statistical summary: \n\n {statistical_summary}\n\n")


df = pd.DataFrame(reHandle)
print("The years listed in the dataset: ")
print(df['List Year'])
city1 = 'Danbury'
city2 = 'Bristol'
city1_data = df.loc[df['Town'] == city1]
city2_data = df.loc[df['Town'] == city2]
mean_city1 = city1_data['Sale Amount'].mean()
mean_city2 = city2_data['Sale Amount'].mean()
min_city1 = city1_data['Sale Amount'].min()
minn_city2 = city2_data['Sale Amount'].min()
max_city1 = city1_data['Sale Amount'].max()
max_city2 = city2_data['Sale Amount'].max()
if mean_city1 > mean_city2:
    print(f"On average, single-family homes are more expensive in {city1}.")
elif mean_city1 < mean_city2:
    print(f"On average, single-family homes are more expensive in {city2}.")
else:
    print("The average sale amounts for single-family homes are the same in both cities.")


```
